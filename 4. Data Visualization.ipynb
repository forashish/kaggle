### Bar Charts and Heatmaps

```python
flightFile = r'C:\Users\singh\Documents\GitHub\kaggle\data\flight_delays.csv'
flightData = pd.read_csv(flightFile, index_col = 'Month')

flightData
```

```python
plt.figure(figsize=(10, 6))
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")
sns.barplot(x=flightData.index, y=flightData['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

### Distributions

```python
irisFile = r'C:\Users\singh\Documents\GitHub\kaggle\data\iris.csv'
irisData = pd.read_csv(irisFile, index_col='Id')

irisData.head()
```

```python
#Histogram
sns.distplot(a=irisData['Petal Length (cm)'], kde=False)
```

```python
# kernel density estimate (KDE) plot
sns.kdeplot(data=irisData['Petal Length (cm)'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=irisData['Petal Length (cm)'], y=irisData['Sepal Width (cm)'], kind='kde')
```

```python
irisSetFile = r'C:\Users\singh\Documents\GitHub\kaggle\data\iris_setosa.csv'
irisVeriFile = r'C:\Users\singh\Documents\GitHub\kaggle\data\iris_versicolor.csv'
irisVirgFile = r'C:\Users\singh\Documents\GitHub\kaggle\data\iris_virginica.csv'

irisSetData = pd.read_csv(irisSetFile, index_col='Id')
irisVeriFile = pd.read_csv(irisVeriFile, index_col='Id')
irisVirgFile = pd.read_csv(irisVirgFile, index_col='Id')
```

```python
# Histogram for each species

sns.distplot(a=irisSetData['Petal Length (cm)'], label='Iris Santosa', kde=False)
sns.distplot(a=irisVeriFile['Petal Length (cm)'], label='Iris Versicolor', kde=False)
sns.distplot(a=irisVirgFile['Petal Length (cm)'], label='Iris Virginica', kde=False)

plt.title('Histogram of Petal Lenght, by Species')
plt.legend()
```

```python
sns.kdeplot(data=irisSetData['Petal Length (cm)'], shade=True)
sns.kdeplot(data=irisVeriFile['Petal Length (cm)'], shade=True)
sns.kdeplot(data=irisVirgFile['Petal Length (cm)'], shade=True)

plt.title("Distribution of Petal Lengths, by Species")
```
